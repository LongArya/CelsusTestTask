# antialiasing EDA

Нужно понять какой минимальный радиус круга можно использовать в датасете, чтобы семпл был валидным. Так как целевые изображения довольно маленькие, скорее всего придется использовать antialiasing (AA). Посмотрим на эффект AA на нескольких значениях радиуса, начиная с минимального

```python
GREEN = (0, 255, 0)


def display_AA_effect_on_circle(radius: int):    
    no_aa_img = np.zeros((IMG_SIZE, IMG_SIZE, 3), dtype=np.uint8)
    aa_img = np.zeros((IMG_SIZE, IMG_SIZE, 3), dtype=np.uint8)

    circle = Circle(x_center_pxl=IMG_SIZE // 2, y_center_pxl=IMG_SIZE // 2, radius_pxl=radius)
    fig, (no_aa_ax, with_aa_ax) = plt.subplots(1, 2)
    plot_filled_circle(no_aa_img, circle, color=GREEN, use_aa=False)
    no_aa_ax.imshow(no_aa_img)
    no_aa_ax.set_title("WITHOUT AA")
    plot_filled_circle(aa_img, circle, color=GREEN, use_aa=True)
    with_aa_ax.imshow(aa_img)
    with_aa_ax.set_title("WITH AA")
    fig.suptitle(f"RADUIS = {radius} pixels")
    plt.show()


display_AA_effect_on_circle(1)
display_AA_effect_on_circle(2)
display_AA_effect_on_circle(3)
display_AA_effect_on_circle(4)
display_AA_effect_on_circle(5)
```
