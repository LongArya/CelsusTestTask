```python
import os 
import sys 
import numpy as np 
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
from src.data.utils import (
    generate_random_circle, 
    plot_filled_circle, 
    generate_random_square, 
    plot_filled_square
)
from src.schemas.data.figures import Circle
from src.consts import IMG_SIZE, MIN_CIRCLE_RADUIS_PXLS, MIN_SQUARE_SIDE_PXLS
import matplotlib.pyplot as plt 

GREEN = (0, 255, 0)
MOSAIC_H, MOSAIC_W = 10, 10
GAP = 1
```

# CIRCLES EXAMPLE

```python
fig, axes = plt.subplots(MOSAIC_H, MOSAIC_W, figsize=(20, 20))
for row_index in range(MOSAIC_H):
    for col_index in range(MOSAIC_W):
        img = np.zeros((IMG_SIZE, IMG_SIZE, 3), dtype=np.uint8)
        circle = generate_random_circle((IMG_SIZE, IMG_SIZE), MIN_CIRCLE_RADUIS_PXLS, GAP)
        plot_filled_circle(img, circle, GREEN, use_aa=True)
        axes[row_index][col_index].imshow(img)
        axes[row_index][col_index].set_xticks([])  
        axes[row_index][col_index].set_yticks([])

fig.tight_layout()
plt.show()
```

# SQUARES EXAMPLE

```python
fig, axes = plt.subplots(MOSAIC_H, MOSAIC_W, figsize=(20, 20))
for row_index in range(MOSAIC_H):
    for col_index in range(MOSAIC_W):
        img = np.zeros((IMG_SIZE, IMG_SIZE, 3), dtype=np.uint8)
        circle = generate_random_square((IMG_SIZE, IMG_SIZE), MIN_SQUARE_SIDE_PXLS, GAP)
        plot_filled_square(img, circle, GREEN, use_aa=True)
        axes[row_index][col_index].imshow(img)
        axes[row_index][col_index].set_xticks([])  
        axes[row_index][col_index].set_yticks([])

fig.tight_layout()
plt.show()
```
