```python
import os 
import sys 
import numpy as np 
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
from src.data.fugure_sample_generation import init_tab10_colors_generator, init_texture_generator
from src.schemas.data.figures import Circle
import matplotlib.pyplot as plt 

MOSAIC_H, MOSAIC_W = 10, 10
```

# COLORED SAMPLES DEMO

```python
colored_generator = init_tab10_colors_generator()

fig, axes = plt.subplots(MOSAIC_H, MOSAIC_W, figsize=(20, 20))
for row_index in range(MOSAIC_H):
    for col_index in range(MOSAIC_W):
        img, meta = colored_generator.generate()
        axes[row_index][col_index].imshow(img)
        axes[row_index][col_index].set_xticks([])  
        axes[row_index][col_index].set_yticks([])

fig.tight_layout()
plt.show()
```
